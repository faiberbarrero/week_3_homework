<p>for an added bonus you can make sure that if a pokemon has 4 moves the user can choose one of them to replace with a new move. </p>

```python
from IPython.display import Image
import requests
class Move_Tutor:
    def __init__(self):
        self.move_list = set()
        self.poke_name = ''
        
    def teach_move(self):
        move_name = input(f'What would you like to teach your {self.poke_name}? ')
        r = requests.get(f"https://pokeapi.co/api/v2/move/{move_name}/")
        if r.status_code == 200:
            moves = r.json()
            if move_name in self.move_list:
                print(f'Your {self.poke_name} already knows {move_name}')
                return
            if len(self.move_list) >= 4:
                move_replace = input(f'Which of the following moves, {self.move_list}, would you like to replace? ')
                if move_replace not in self.move_list:
                    print(f'{self.poke_name} doesn\'t know {move_replace} yet!' )
                    return
                self.move_list.remove(move_replace)
                print(f'{self.poke_name} forgot {move_replace}')
            self.move_list.add(move_name)
            print(f'Your {self.poke_name} just learned {move_name}!!')
        else:
            print(f"Ran into an issue {r.status_code}")
            return
        
    def show_moves(self):
        for move in self.move_list:
            print(move)
        

class Pokemon(Move_Tutor):
    def __init__(self,name):
        super().__init__()
        self.name = name
        self.poke_name = name
        self.types = []
        self.abilities = {}
        self.weight = None
        self.poke_api_call()
        
    def poke_api_call(self):
        r = requests.get(f'https://pokeapi.co/api/v2/pokemon/{self.name}')
        if r.status_code == 200:
            pokemon = r.json()
        else:
            print(f'Ran into an issue {r.status_code}')
            return
        self.name = pokemon['name']
        self.types = [poke['type']['name'] for poke in pokemon['types']]
        self.abilities = [poke['ability']['name'] for poke in pokemon['abilities']]
        self.weight =  pokemon['weight']
        print(f"{self.name}'s data has been updated")
       
        

    def display(self):
        r = requests.get(f'https://pokeapi.co/api/v2/pokemon/{self.name}')
        if r.status_code == 200:
            pokemon = r.json()
        else:
            print(f'Ran into an issue {r.status_code}')
            return

        self.img = pokemon['sprites']['other']['official-artwork']['front_default']
        display(Image(str(self.img), width = 300))
        
    def __repr__(self):
        return f'You caught a {self.name}'
            
```
